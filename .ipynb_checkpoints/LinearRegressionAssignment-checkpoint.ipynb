# Multiple Linear Regression Case Study
 - Prediction of Demand for Shared Bikes

#### **Problem Statement**
  A US bike-sharing provider **BoomBikes** aspires to understand the factors on which the demand for the shared bikes depends.
  Bikes are made available for shared use to individuals on a short term basis for a price or free. 

  The company wants to know - 
  - Which variables are significant in predicting the demand for shared bikes.
  - How well those variables describe the bike demands

#### **Business Goal**
Prepare a model keeping the demand for shared bikes as target variable from the available independent variables. 
It will be used by the management to understand how exactly the demands vary with different features. 
They can accordingly manipulate the business strategy to meet the demand levels and meet the customer's expectations. 
Further, the model will be a good way for management to understand the demand dynamics of a new market.

## Step 4: Splitting the Data into Training and Testing Sets

```python
# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(bikesharingdata, train_size = 0.70, random_state = 42)
```

```python
print(df_train.shape)
print(df_test.shape)
```

#### Rescaling the Features

```python
# we will use MinMax scaling to rescale the variables so that they have a comparable scale, so that all the data 
# converges within the range of 0 to 1
```

```python
# instantiate an object
scaler = MinMaxScaler()

num_vars = ['temp','atemp','hum','windspeed','cnt']

#Fit and transform training set only
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_test[num_vars] = scaler.transform(df_test[num_vars])
```

```python
df_train.describe()
```

```python
df_train.head(2)
```

#### Dividing into X and Y sets for the model building

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

## Step 5: Building a linear model

```python
def calculateVIF(X_train):
    vif_data=pd.DataFrame()
    vif_data["Feature"]=X_train.columns
    vif_data["VIF"]=[variance_inflation_factor(X_train.values,i) 
                     for i in range(len(X_train.columns))]
    return vif_data
```

### Model : Linear Regression Model with All Independent Variables

```python
X_train_sm = sm.add_constant(X_train)
```

```python
lr = sm.OLS(y_train,  X_train_sm).fit()
print(lr.summary())
```

```python
vif_data = calculateVIF(X_train_sm)
vif_data
```

#### Build model using RFE - Recursive feature elimination - by selection of the most relevant features

```python
lm = LinearRegression()
rfe = RFE(estimator=lm,n_features_to_select=15)
rfe = rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
selected_col = X_train.columns[rfe.support_]
selected_col
```

### Model 1: using RFE technique

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[selected_col]
```

```python
# Adding a constant variable 
X_train_sm = sm.add_constant(X_train_rfe)

# Running the linear model
lm = sm.OLS(y_train,X_train_sm).fit()  

# Print the summary of our linear model
print(lm.summary())
```

**atemp** is insignificant in presence of other variables and the variable can be dropped

```python
X_train_new = X_train_sm.drop(["atemp"], axis = 1)
```

### Model 2 - dropping atemp

```python
model_2 = sm.OLS(y_train,X_train_new).fit() 
print(model_2.summary())
```

**weekday_sat** is insignificant in presence of other variables and the variable can be dropped

```python
vif_data = calculateVIF(X_train_new)
vif_data
```

```python
X_train_new = X_train_new.drop(["weekday_sat"], axis = 1)
```

### Model 3 - dropping weekday_sat

```python
model_3 = sm.OLS(y_train,X_train_new).fit() 
print(model_3.summary())
```

```python
vif_data = calculateVIF(X_train_new)
vif_data
```

Based on the provided regression summary, **mnth_dec** appears to be the least significant predictor, with a coefficient of -0.0629

```python
X_train_new = X_train_new.drop(["mnth_dec"], axis = 1)
```

### Model 4 : dropping mnth_dec

```python
model_4 = sm.OLS(y_train,X_train_new).fit() 
print(model_4.summary())
```

Based on the provided regression summary, mnth_jul appears to be the least significant predictor, with a coefficient of -0.0636

```python
vif_data = calculateVIF(X_train_new)
vif_data
```

```python
X_train_new = X_train_new.drop(["mnth_jul"], axis = 1)
```

### Model 5 : dropping mnth_jul

```python
model_5 = sm.OLS(y_train,X_train_new).fit() 
print(model_5.summary())
```

Based on the provided regression summary, mnth_nov appears to be the least significant predictor, with a coefficient of -0.0683

```python
vif_data = calculateVIF(X_train_new)
vif_data
```

```python
X_train_new = X_train_new.drop(["mnth_nov"], axis = 1)
```

#### Model 6 : dropping mnth_nov

```python
model_6 = sm.OLS(y_train,X_train_new).fit() 
print(model_6.summary())
```

```python
vif_data = calculateVIF(X_train_new)
vif_data
```

Based on the provided regression summary, here's an assessment of the model's performance:

**R-squared and adjusted R-squared**: Both values are relatively high (0.824 and 0.821, respectively), indicating that the model explains a substantial portion of the variance in the target variable.

**Individual coefficients:** All coefficients have p-values less than 0.05, indicating that each predictor is statistically significant at the 5% level.

**VIF values:** VIF values are less than 5, suggests that multicollinearity is not a major issue.

## Step 6: Residual Analysis of the train data

```python
y_train_pred = model_6.predict(X_train_new)
```

```python
res = y_train - y_train_pred
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot(res, bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  
plt.xlabel('Errors', fontsize = 18)    
```

Errors are normally distribured with mean 0. 

### Q-Q Plot 

```python
sm.qqplot(res)
plt.title('Q-Q Plot of Residuals')
plt.show()
```


### Step 7: Making Predictions Using the Final Model

```python
y_test = df_test.pop('cnt')
X_test = df_test
```

```python
X_train_new = X_train_new.drop(['const'], axis=1)
```

```python
X_test_new = X_test[X_train_new.columns]

# Adding a constant variable 
X_test_new = sm.add_constant(X_test_new)
# Making predictions
y_test_pred = model_6.predict(X_test_new)
```

## Step 8: Model Evaluation

```python
## r2 values for train and test data
```

```python
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

print("Train:",round(r2_train,3),"Test:",round(r2_test,3))
```

The R-squared values for the training and test sets are similar (0.824 and 0.823, respectively), This suggests that the model is generalizing well to unseen data and is not overfitting.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Calculate MSE
mse = mean_squared_error(y_true=y_test, y_pred=y_test_pred)

# Calculate MAE
mae = mean_absolute_error(y_true=y_test, y_pred=y_test_pred)

# Calculate RMSE
rmse = mean_squared_error(y_true=y_test, y_pred=y_test_pred, squared=False)

print("MSE:", mse)
print("MAE:", mae)
print("RMSE:", rmse)
```

These metrics indicate that the model is performing well.


```python
# Plotting y_test and y_test_pred to understand the spread.

fig = plt.figure()
plt.scatter(y_test,y_test_pred)
fig.suptitle('y_test vs y_test_pred', fontsize=20)              # Plot heading 
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_test_pred', fontsize=16)  
```

We can observe that variance of the residuals (error terms) is constant across predictions. i.e error term does not vary much as the value of the predictor variable changes.

```python
## Based on the coefficients provided, the linear regression equation for predicting bike demand can be written as

Bike sharing Demand = 0.2358 * yr + 0.4371* temp − 0.1535 * hum − 0.1131 * windspeed − 0.1570 * season_spring + 0.0610 * season_winter +
              0.0640 * mnth_mar + 0.0752 * mnth_sep − 0.0552 * weathersit_2 − 0.2256 * weathersit_3 + 0.3173
```

Based on the provided regression summary, here's an interpretation of the coefficients for each predictor:

1. **Year (yr):**
For each increase of one year, the bike sharing demand is expected to increase by 0.2358 units, holding other variables constant. This suggests a positive trend in bike sharing demand over time.

2. **Temperature (temp):**
For each unit increase in temperature, the bike sharing demand is expected to increase by 0.4371 units, holding other variables constant. This indicates a strong positive relationship between temperature and bike sharing demand.

3. **Humidity (hum):**
For each unit increase in humidity, the bike sharing demand is expected to decrease by 0.1535 units, holding other variables constant. This suggests a negative relationship between humidity and bike sharing demand.

4. **Windspeed:**
For each unit increase in windspeed, the bike sharing demand is expected to decrease by 0.1131 units, holding other variables constant. This suggests a negative relationship between windspeed and bike sharing demand.

5. **Season (season_spring, season_winter):**
Demand is significantly lower in the spring and Winter.

6. **Month (mnth_mar, mnth_sep):**
Demand is slightly higher in March and September.

7. **Weather Situation (weathersit_2, weathersit_3):**
A slight decrease in weathersit_2 and weathersit_3, suggesting a substantial decrease in demand.
